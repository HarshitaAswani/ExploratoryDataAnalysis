```python
df['Order Date'] = df['Order Date'].apply(pd.to_datetime)
df['Ship Date'] = df['Ship Date'].apply(pd.to_datetime)
```

```python
df['Region'] = df['Region'].astype(str)
df['Country'] = df['Country'].astype(str)
df['Item Type'] = df['Item Type'].astype(str)
df['Sales Channel'] = df['Sales Channel'].astype(str)
df['Order Priority'] = df['Order Priority'].astype(str)
```

```python
df.info()
```

```python
# Create a pairplot to visualize the relationships between variables
sns.pairplot(df, vars=['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Profit'])
```

```python
# Create a heatmap to visualize the correlation between variables
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Calculate average revenue per order and print it
avg_revenue = df.groupby('Order ID')['Unit Price'].sum().mean()
print('Average revenue per order: ${:.2f}'.format(avg_revenue))
```

```python
# Plot a histogram of order quantity
plt.hist(df['Units Sold'], color='red')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.title('Distribution of order quantity')
plt.show()
```

```python
# Plot a line chart of yearly revenue
daily_revenue = df.groupby('Ship Date')['Unit Price'].sum()
plt.plot(daily_revenue, color='green')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Yearly revenue')
plt.show()
```

```python
# Calculate top-selling products and print them
top_products = df.groupby('Item Type')['Units Sold'].sum().sort_values(ascending=False)
print('Top-selling products:')
print(top_products.head())
```

```python
# Calculate total revenue by region and print it
revenue_by_region = df.groupby('Region')['Unit Price'].sum()
print('Revenue by region:')
print(revenue_by_region)
```

```python
# Plot a pie chart of revenue by region
plt.pie(revenue_by_region, labels=revenue_by_region.index, autopct='%1.1f%%')
plt.title('Revenue by region')
plt.show()
```

```python
# Calculate total revenue by product category and print it
revenue_by_category = df.groupby('Item Type')['Unit Price'].sum()
print('Revenue by product category:')
print(revenue_by_category)
```

```python
# Plot a bar chart of revenue by product category
revenue_by_category.plot(kind='bar', color='orange')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.title('Revenue by product category')
plt.show()
```

```python
# Calculate total revenue by country and print it
revenue_by_country = df.groupby('Country')['Unit Price'].sum().sort_values(ascending=False)
print('Revenue by country:')
print(revenue_by_country.head())
```

```python
# Plot a bar chart of revenue by country
revenue_by_country.plot(kind='bar', color='purple', figsize=(14,5))
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.title('Revenue by country')
plt.show()
```

```python
# Calculate the average revenue per order by country and print it
avg_revenue_by_country = df.groupby('Country')['Unit Price'].mean().sort_values(ascending=False)
print('Average revenue per order by country:')
print(avg_revenue_by_country.head())
```

```python
# Calculate total revenue by channel type and print it
revenue_by_channel = df.groupby('Sales Channel')['Unit Price'].sum().sort_values(ascending=False)
print('Revenue by channel type:')
print(revenue_by_channel)
```

```python
# Plot a bar chart of revenue by channel type
revenue_by_channel.plot(kind='pie')
plt.xlabel('Channel Type')
plt.ylabel('Revenue')
plt.title('Revenue by channel type')
plt.show()
```

```python
# Calculate the average revenue per order by channel type and print it
avg_revenue_by_channel = df.groupby('Sales Channel')['Unit Price'].mean().sort_values(ascending=False)
print('Average revenue per order by channel type:')
print(avg_revenue_by_channel)
```
